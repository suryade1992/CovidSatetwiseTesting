```python
import pandas as pd
data_df = pd.read_excel('/Users/sde24/Downloads/Covid19.xlsx')
data_df = data_df[data_df['Total Tested']>=0].sort_values(['State','Updated On']).drop_duplicates('State',keep='last')
data_df = data_df[['State', 'Total Tested', 'Positive']]
population_df = pd.read_excel('/Users/sde24/Downloads/Population.xlsx').rename(columns={'State or union territory':'State','Population (2020 estimates)':'Population'})
chart_df = data_df.merge(population_df,on='State',how='left')
```

```python
import math
# sk = pd.DataFrame.from_records([{'State':'South Korea','Total Tested':546463,'Positive':10653,'Population':51600000}])
# chart_df = chart_df.append(sk,ignore_index=True)
chart_df['positive_rate'] = chart_df.apply(lambda x: 15000*x['Positive']/x['Total Tested'],axis=1)
# chart_df['Total Tested (log)'] = chart_df.apply(lambda x: math.log(x['Total Tested'],10),axis=1)
# chart_df['Population (log)'] = chart_df.apply(lambda x: math.log(x['Population'],10),axis=1)
chart_df.sample(2)
```

```python
import seaborn as sns
sns.scatterplot(x="Population (log)", y="Total Tested (log)", data=chart_df)
```

```python
chart_df1 = chart_df[chart_df.Population>=8000000]
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
y = chart_df1['Total Tested'].values
x = chart_df1["Population"].values
n = chart_df1['State'].values
s = chart_df1['positive_rate'].values

fig, ax = plt.subplots()
plt.scatter(x, y,s=s)
plt.gcf().set_size_inches((15, 10))  

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
    
slope = 1/10000
c = 0

x_min, x_max = ax.get_xlim()
y_min, y_max = x_min*slope, x_max*slope
ax.plot([x_min, x_max], [y_min, y_max],c='r', label='1 in 10k')
ax.set_xlim([0, x_max])

slope = 10/10000
c = 0

y_min, y_max = ax.get_ylim()
x_min, x_max = y_min/slope, y_max/slope
ax.plot([x_min, x_max], [y_min, y_max],c='y', label='1 in 1k')
ax.set_ylim([0, y_max])

slope = 546463/51600000
c = 0

y_min, y_max = ax.get_ylim()
x_min, x_max = y_min/slope, y_max/slope
ax.plot([x_min, x_max], [y_min, y_max],c='g', label='South Korea')
ax.set_ylim([0, y_max])
ax.legend()
ax.set_title('State wise testing rate: Radius proportional to PositiveCases/TotalTest')
plt.savefig('CovidIndiaAnalysis_Source_covid19india_org.jpeg')
```
